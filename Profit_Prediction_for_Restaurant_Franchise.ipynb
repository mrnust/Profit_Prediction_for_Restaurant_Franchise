```python
# Importing necessary libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error

# Load the dataset
data = pd.read_csv('/content/res_profit_data.csv')

# Visualize the dataset
plt.scatter(data['population'], data['profit'])
plt.xlabel('population')
plt.ylabel('profit')
plt.title('Restaurant Profit vs. Population')
plt.show()

# Prepare the data for modeling
X = data['population'].values.reshape(-1, 1)  # Feature (Population)
y = data['profit'].values  # Target variable

# Create and train the linear regression model
model = LinearRegression()
model.fit(X, y)

# Make predictions on the training set
y_pred = model.predict(X)

# Visualize the best-fit line
plt.scatter(X, y, color='blue')
plt.plot(X, y_pred, color='red', linewidth=2)
plt.xlabel('population')
plt.ylabel('profit')
plt.title('Linear Regression: Restaurant Profit vs. Population')
plt.show()

# Calculate the accuracy of the model
mse = mean_squared_error(y, y_pred)
print("Mean Squared Error:", mse)
```
